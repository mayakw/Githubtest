# If - ElIf - Else

<div>

</div>

<br />

Python Format:

```
if condition1:
    #code that runs if condition1 is 'True'
elif condition2:
    #code that runs if condition1 is 'False' AND condition 2 is 'True'
.
.
.
elif conditionN:
    #code that runs if ALL previous conditions are 'False' AND conditionN is 'True'
.
.
.
else:
    #code that runs if All previous condition are 'False'
```

````python
condition1 = True
condition2 = True

#condition1 = False
#condition2 = False

#condition1 = True
#condition2 = False

#condition1 = False
#condition2 = True

if condition1:
    print('condition1 Met')
elif condition2:
    print('condition1 was NOT met, condition2 met')
else:
    print('neither condition1 nor condition2 were met')
    
print('after if statement')
````

# Nested Ifs

````python
cond1 = False
cond2 = True

if cond1:
    print('if-beginning')
    if cond2:
        print('if-if')
        print('cond1 True AND cond2 True')
    else:
        print('if-else')
        print('cond1 True, but not cond2')
    print('if-end')
else:
    print('else')

print('after if')
````

````python
#SIRS (systemic inflammatory response syndrome) criteria -> used to identify septic patients
hr = 92 #beats/min
temp = 39 #deg celsius

if hr > 90:
    if temp > 38 or temp < 36:
        print('Risk for SIRS')
    else:
        print('Minimal Risk for SIRS')
else:
    print('no risk for SIRS')
````

#### if-else vs. if-if

Why not just use 2 if statements??

````python
condition1 = True
condition2 = True

#condition1 = False
#condition2 = False

#condition1 = True
#condition2 = False

#condition1 = False
#condition2 = True

if condition1:
    print('condition1 Met')
if condition2:
    print('condition2 met')
else:
    print('condition2 NOT met')
    
print('after if statement')
````

***This is where you have to think about your program and what real-world scenario you're trying to model***

* In an if-if statement, both are independent from each other -> both can be run
* In an if-else statement, they are dependent -> only one can be run

````python
#SIRS (systemic inflammatory response syndrome) criteria -> used to identify septic patients
hr = 92 #beats/min
temp = 39 #deg celsius

if hr > 90:
    if temp > 38 or temp < 36:
        print('Risk for SIRS')
    else:
        print('Minimal Risk for SIRS')
else:
    print('no risk for SIRS')
````
