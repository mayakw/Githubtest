# Tuples
- They are similar to lists, but they are immutable. 

> What does immutable mean? 
>>Once an element is inside a tuple, it cannot be reassigned. 

- Tuples use $paranthesis$: (1,2,3). While lists use $square brackets$: [1,2,3]

```python
t = (1,2,3) #creating a tuple variable
type(t) #check the type of variable t
```

```python
l = [1,2,3] #creating a list variable
type(l) #checking the type of variable l
```

- Since tuples are immutable (unchangable), you can never add more elements in a tuple

```python
t[2]=10 #trying to change the value in tuple (gives an error, since tuples are immutable)
print(t)
```

```python
l[2]=10 #Oh! we can change the value easily in a list
print(l)
```

- Tuple allows to store various object types just like lists

```python
tup = ('one', 2)
print(tup)
```

- We can also use slicing and indexing just like list

```python
tup[0] #this gives us the first item
```

```python
tup[-1] #just like lists, we can start from the last item as well
```

- We can add 2 tuples together to form a new tuple

```python
added_tuple = t + tup #adding the two tuples using '+' sign
print(added_tuple)
```

- There are 2 methods in tuples, count() and index()
- Methods are specific functions that are associated with object/data type. 

```python
c = ('a', 'a', 'b', 'c') #create a tuple with duplicate entries
print(c)
```

```python
c.count('a') #count the total number of times 'a' has been repeated
```

```python
c.index('a') #shows the location when 'a' first appeared 
```
