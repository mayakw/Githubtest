## Melt

* Melt is used to reshape/transform a DataFrame.
* Reshapes from a 'wide' format to a 'long' format.
* AKA 'unpivoting' the data.
* This function is useful to massage a DataFrame into a format where one or more columns are identifier variables (id_vars), while all other columns, considered measured variables (value_vars), are “unpivoted” to the row axis

```python
# Create synthetic patient data
n = 10  # Number of patients
data = {
    'Patient_ID': range(1, n + 1),
    'Name': [''.join(random.choice(string.ascii_uppercase) for _ in range(3)) for _ in range(n)],
    'Age': [random.randint(18, 90) for _ in range(n)],
    'Gender': [random.choice(['Male', 'Female']) for _ in range(n)],
    'Blood_Type': [random.choice(['A', 'B', 'AB', 'O']) for _ in range(n)]
}

# Create synthetic medical data
data['Heart_Rate'] = [random.randint(60, 100) for _ in range(n)]
data['Blood_Pressure_Systolic'] = [random.randint(90, 160) for _ in range(n)]
data['Blood_Pressure_Diastolic'] = [random.randint(60, 100) for _ in range(n)]
data['Temperature'] = [round(random.uniform(36.0, 37.5), 1) for _ in range(n)]

# Create the DataFrame
df = pd.DataFrame(data)

# Print the DataFrame
df
```

```python
columns_to_melt = ['Patient_ID', 'Name', 'Age', 'Gender', 'Blood_Type']

# Use melt to reshape the DataFrame
df_melted = df.melt(id_vars=columns_to_melt, var_name='Measurement', value_name='Value')
df_melted
```

```python
import seaborn as sns

sns.barplot(data = df_melted, x='Gender', y='Value', hue = 'Measurement')
```

```python
sns.barplot(data = df, x='Gender', y='Heart_Rate', hue = 'Blood_Type')
```

## Pivot

* Also used to reshape (or Pivot) the DF
* Converts unique values in one column into multiple columns
* Opposite of what 'melt' does

```python
df_pivoted = df_melted.pivot(index=columns_to_melt, columns=['Measurement'], values=['Value'])
df_pivoted.reset_index(inplace=True)

df_pivoted
```

```python
df_pivoted.columns = ['Patient_ID', 'Name', 'Age', 'Gender', 'Blood_Type', 'Heart_Rate', 'Blood_Pressure_Systolic', 'Blood_Pressure_Diastolic', 'Temperature']

df_pivoted
```

## Aggregations

#### Aggregation methods allow you to do Basic Statistical Operations

```python
# Average age
df['Age'].mean()
```

```python
# Maximum Heart Rate
df['Heart_Rate'].max() #.min() .std()
```

## Grouped Aggregations with groupby()

```python
# Groupby Gender -> get mean age
df.groupby('Gender')['Age'].mean()
```

```python
# Groupby Gender and Blood Type -> get max HR
df.groupby(['Gender', 'Blood_Type'])['Heart_Rate'].max()
```

## Multiple Aggregations with agg()

```python
# .agg() be careful about the methods
df.aggregate(['mean', 'min', 'max', 'sum'])
df.aggregate(['mean', 'min', 'max', 'sum'])
```

```python
df[['Heart_Rate','Heart_Rate']].agg(['mean', 'min', 'max', 'sum']) # input to agg is a list
```

```python
# Combine .groupby() + .agg() input to agg is dictionary
df.groupby('Gender').agg({
    'Age': ['mean', 'std'],
    'Temperature': ['min', 'max']
})
```

```python
df.groupby(['Gender'], as_index=False).agg(
                {'Heart_Rate':'mean', 'Temperature':'mean', 'Blood_Pressure_Systolic': 'max'
            })
```

## Applying Custom Aggregations with apply()

```python
def some_func(x):
    return x**2
df.groupby(['Gender'], as_index=False)['Blood_Pressure_Systolic'].apply(some_func)
```

```python
# Define the function to create a combined blood pressure column
df['Blood_Pressure'] = df.apply(lambda x: f"{x['Blood_Pressure_Systolic']}/{x['Blood_Pressure_Diastolic']}", axis=1)

# Now group by 'Gender' and aggregate the combined values as a list
df.groupby(['Gender'], as_index=False).agg({'Blood_Pressure': list})
```

```python
%%time

#adding new column
temp_f = []

for index,row in df.iterrows():
    temp_f.append(9/5 * row['Temperature'] + 32)
    
df['Temperature_F'] = temp_f
df
```

CPU times:

User: The time spent by the CPU running your Python code (or user-level code). Here, it’s 1.06 ms. This represents the amount of time the CPU was actively executing instructions in user mode.
Sys: The time spent in system-level operations (such as memory management and other system calls) by the CPU. Here, it’s 1.2 micros. This represents the overhead of running the program, such as interacting with the operating system.
Together, these two times (1.2ms) represent the total CPU time.

Wall time: This is the actual elapsed (or "wall-clock") time taken for the entire code block to run, as if measured by a stopwatch. In this case, it’s 1.07 ms. Wall time includes all time spent in waiting states, such as time spent waiting for I/O operations or any delays due to multi-threading or multi-processing.

The difference between CPU time (6.76 s) and Wall time suggests that there may be additional overhead, such as waiting for I/O.

```python
#another example with very large df
data = np.random.rand(10000, 10000) #np array
dummy_df = pd.DataFrame(data) #pd df
```

```python
%%time

#df sum
result = dummy_df.sum().sum()
```

```python
%%time

#adding new column
summ = 0

for index,row in dummy_df.iterrows():
    summ += sum(row)
```

```python
%%time

#np sum
result = np.sum(data)
```

```python
%%time

#adding new column
summ = 0

for index,row in tqdm(dummy_df.iterrows()):
    summ += sum(row)
```

```python
%%time

a = 2
for i in tqdm(range(1000000)):
    a *= 2
```

tqdm(range(1000000)) wraps the range(1000000) iterable with a progress bar.
As the loop iterates, tqdm displays a progress bar.
The progress bar provides real-time feedback on the iteration count, completion percentage, and estimated time remaining.
