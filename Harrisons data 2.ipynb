```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('hp_data2.csv')
df.head()
```

```python
Al = df['Al2O3/SiO2'].tolist()
Mg = df['MgO/SiO2'].tolist()
print (Al, Mg)
```

```python
x1, y1 = 0, 1.12
x2, y2 = 0.02, 1.057

arfont = {'fontname':'Arial'}
plt.ylim(0.8, 1.2)
plt.xlim(0,0.02)
plt.xlabel('$Al_{2}O_{3}/SiO_{2}$',**arfont)
plt.ylabel('$MgO/SiO_{2}$',**arfont)

plt.plot([x1, x2], [y1, y2], color = 'black')
plt.scatter(Al,Mg, color = '#5b7c99')

plt.savefig('partial_melt_fractionation.png', dpi=300)
plt.show()
```
