```python
import numpy as np
import pandas as pd
```

```python
#importing the data
df = pd.read_csv("breast-cancer.data",header = None, sep =',')
df.columns = ['recurrence', 'age', 'menopause', 'tumor_size', 'inv_nodes', 'node_caps', 'deg_malig', 'breast', 'breast_quad', 'irradiat']
df['id'] = range(1, len(df) + 1)
df.head()
```

```python
#Melt
tumor_melt = ['recurrence', 'age', 'menopause','node_caps', 'deg_malig','breast', 'breast_quad','irradiat','id'] #selecting all the comuns I want to keep

df_melt = df.melt(id_vars=tumor_melt, var_name='Tumor', value_name='Value')
df_melt.head()
```

```python
#Pivot
df_pivot = df_melt.pivot(index= tumor_melt, columns=['Tumor'], values=['Value'])
df_pivot.head()
```

```python
#Aggregation
df.aggregate({"deg_malig":['min','max','mean']}) #deg_malig is the only column with numerical values
```

```python
#Iteration
for ind in df.index:
    print (df['breast_quad'][ind],' of ',df['breast'][ind],' breast')  #Iterate through the table to return the location of the cancer
```

```python
#Groupby
df.groupby('age')['deg_malig'].mean() #return the average deg_malig for each age group
```
