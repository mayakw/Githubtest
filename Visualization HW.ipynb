```python
#%% libraries
import pandas as pd
import matplotlib.pyplot as plt
#%% data

url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
covid_df = pd.read_csv(url, index_col=0)
```

```python
covid_df.head()
```

```python
#Question 1
'''
Create a visualization that shows all of the counties in Utah as a time series,
similar to the one shown in slide 22 during the lecture. The graphic should
-Show cases over time
-Have all counties plotted in a background color (something like grey)
-Have a single county plotted in a contrasting color (something not grey)
-Have well formatted dates as the X axis
'''
#pulling the data I want- only rowns in Utah
covid_ut = covid_df.loc[covid_df['Province_State']=='Utah']
#cleaning all the columns with data I don't need
covid_ut_clean = covid_ut.drop(['iso2','iso3','code3','FIPS','Province_State','Country_Region','Lat','Long_','Combined_Key'],axis  = 1)
#Transposing the table in a way that makes more sense to graph in my head
ut_data = covid_ut_clean.transpose()
ut_data.columns = ut_data.iloc[0] 
ut_data = ut_data[1:]

#removing thge data that is not actaully a county in Utah
ut_data = ut_data.drop(['Bear River','Central Utah','Out of UT','Southeast Utah','Southwest Utah','TriCounty','Unassigned','Weber-Morgan'],axis=1)
```

```python
#plotting all the lines
ut_data.plot(color='grey',legend=False)
#Making one of them red
ut_data['Salt Lake'].plot(color = 'red')
#adding labels
plt.title("Total COVID cases in Salt Lake County over time relative to other counties")
plt.xlabel("Date")
plt.ylabel("Number of cases")
plt.show()
```

```python
#Question 2
'''
Create a visualization that shows the contrast between the county in Utah with
the most cases to date to a county in Florida with the most cases to date.
The graphic should:
-Have only two counties plotted
-Highlight the difference between the two comparison counties
You may use any style of graphic you like as long as it is effective (dense)
and readable
'''
#pulling the data for Florida
covid_fl = covid_df.loc[covid_df['Province_State']=='Florida']
#covid_fl.head()
#creating a list of the maximum values in each state
fl_max = covid_fl['3/4/23'].max().tolist()
ut_max = covid_ut['3/4/23'].max().tolist()
max_value=[fl_max,ut_max]
print (max_value)

#finding the index of the max values
fl_max_index = covid_fl.index[covid_fl['3/4/23']==fl_max].tolist()
ut_max_index = covid_ut.index[covid_ut['3/4/23']==ut_max].tolist()

#ut_max_index = covid_ut[['3/4/23']].idxmax()
print (fl_max_index, 'this is index')

#getting the county that the max value came from
max_ct = covid_fl.loc[fl_max_index, 'Admin2'].tolist()
max_ct.extend(covid_ut.loc[ut_max_index, 'Admin2'].tolist())
print (max_ct)
```

```python
fig = plt.figure(figsize = (2,3))

plt.bar(max_ct,max_value, width = 0.4)
plt.xlabel('Counties')
plt.ylabel('Number of Cases')
plt.show()
```

```python
#Question 3
#pulling the data from salt lake county
slc_data = ut_data['Salt Lake']
slc_df = slc_data.to_frame()
slc_df = slc_df.reset_index()

#making a list that has the new cases for each day
daily = []
i = 1 #staring on 1, the equation wont work on the first day
while i <= 1142:
    x = slc_df.loc[i,'Salt Lake']
    y = slc_df.loc[i-1,'Salt Lake']
    z = x-y
    daily.append(z)
    i+=1

#adding in 0 new cases for day 1
daily.insert(0,0)
#adding a column for daily cases
slc_df['New Cases'] = daily
#setting the date column type to datetime
slc_df['index'] =slc_df['index'].astype('datetime64[ns]')
```

```python
#plotting the chart
fig, ax1 = plt.subplots(figsize=(10, 5))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Total Cases', color=color)
ax1.plot(slc_df['index'],slc_df['Salt Lake'], color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, 450000)

ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('New Cases', color=color) 
ax2.bar(slc_df['index'],slc_df['New Cases'],color=color, width = 5)
ax2.tick_params(axis='y', labelcolor=color)

plt.title("Running Total and Daily New COVID Cases Salt Lake County")
plt.show()
```

```python
#Question 4
'''
Create a visualization that shows a stacked bar chart of county contributions
to a given state's total cases. You may choose any state (or states).
(https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_stacked.html#sphx-glr-gallery-lines-bars-and-markers-bar-stacked-py)
The graphic should:
-Have a single column delineate a state
-Have each 'slice' or column compontent represent a county
'''
#pulling the data froma single date in Utah, removing counties with no data, and transposing the data
final_count = covid_ut_clean[['Admin2','3/9/23']]
final_count = final_count.where(final_count['3/9/23']>0)
final_count = final_count.transpose()
final_count.columns = final_count.iloc[0]
final_count = final_count[1:]
final_count = final_count.reset_index()
final_count.dropna(how='all', axis=1, inplace=True) 

final_count.head()
```

```python
final_count.plot(x='index', kind='bar', stacked=True,
        title='Total COVID Cases by County', width = 0.1)
plt.xlabel('')
plt.ylabel('Total Cases')
plt.legend(title = "Counties")
plt.show()
```
